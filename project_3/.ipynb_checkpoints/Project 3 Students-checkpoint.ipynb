## 2. Data Preprocessing and Cleaning

```python
# drop datetime info
# -----------
chicago_inspections_2011_to_2013 = chicago_inspections_2011_to_2013.dropna().drop(['Inspection_Date',
                                                                                   'minDate',
                                                                                   'maxDate',
                                                                                   'Inspection_MonthYear'],
                                                                                  axis = 1)
```

```python
# process target 
# -----------
y = chicago_inspections_2011_to_2013['Results']

# decide if you want to binarize the outcome variable 
# -----------
# comment out the following lines of code if you don't want to binarize the target variable
y = y.replace({'Pass w/ Conditions': 'Pass'})
lb_style = LabelBinarizer()
y = lb_style.fit_transform(y)

# recode 0s and 1s so 1s are "Fail"
y = np.where(y == 1, 0 ,1)


# process features
# -----------

# create feature dataset
X = chicago_inspections_2011_to_2013.drop(columns = ['Results', 
                                                     'Facility_Type', 
                                                     'LICENSE_DESCRIPTION', 
                                                     'WARD_PRECINCT'])


# get dummies
X = pd.get_dummies(X)
```

```python
# view feature datset
X.head()
```

```python
# training and test split 
X_train, X_val, y_train, y_val = train_test_split(X, 
                                                  y, 
                                                  train_size = 0.8, 
                                                  test_size=0.2)
```

### Model 1: Random forest

```python
# initialize a random forest classifier
# ----------
rf_classifier = RandomForestClassifier(random_state = 10)             # random seed
```

```python
rf_classifier.get_params()
```

```python
# specify params to search
param_grid = {'n_estimators': [50, 100, 200],
              'max_depth': [None, 10, 20],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [1, 2, 4],
              'max_features': ['sqrt', 'log2']}

# object to search grid
rf_grid_search = GridSearchCV(rf_classifier,
                              param_grid, 
                              cv=3)

rf_grid_search.fit(X_train, y_train)
```

```python
best_rf = rf_grid_search.best_estimator_
print("Best Parameters:", best_rf.get_params())
```

```python
# create feature importance dataframe
feat_importances = pd.concat([pd.DataFrame(X_train.columns),pd.DataFrame(np.transpose(best_rf.feature_importances_))], axis = 1)
feat_importances.columns = ["Feature", "Importance"]

# plot 
sns.barplot(x = "Importance", 
            y = "Feature", 
            data = feat_importances) # identify the 10 most important features
plt.show()
```

```python
# refit a basic decision tree using reduced number of features 
# ----------
rf_reduced_classifier = RandomForestClassifier(# best parameters from CV search
                                               bootstrap=True,
                                               criterion='gini',
                                               ccp_alpha=0.0, 
                                               class_weight=None,
                                               n_estimators=200,
                                               n_jobs=None,
                                               oob_score=False,
                                               max_depth=10,                   # how deep tree nodes can go
                                               min_samples_split=4,           # samples needed to split node
                                               min_samples_leaf=10,            # samples needed for a leaf
                                               min_weight_fraction_leaf=0.0,  # weight of samples needed for a node
                                               max_features='sqrt',             # number of features to look for when splitting
                                               max_leaf_nodes=None,           # max nodes
                                               max_samples=None,
                                               min_impurity_decrease=0.0,   # early stopping
                                               verbose=0,
                                               warm_start=False,
                                               random_state = 10)             # random seed

# pull out the most features
important_features = feat_importances.nlargest(10, 'Importance')['Feature']

# create new dataset with only most important features
X_train_reduced = X_train[X_train.columns[X_train.columns.isin(important_features)]]

# fit the model on the new reduced model
rf_reduced_classifier.fit(X_train_reduced,
                          y_train)

# set column names as list
reduced_column_names = X_train_reduced.columns.tolist()
```

###  Model 2

```python
# initialize an XGBoost classifier
# ----------
# !pip install xgboost
import xgboost as xgb

xgb_classifier = xgb.XGBClassifier(random_state=10)

# define the scoring metrics
scoring = {
          'accuracy': make_scorer(accuracy_score),
          'recall': make_scorer(recall_score),
          'precision': make_scorer(precision_score),
          'f1': make_scorer(f1_score)
          }


# perform cross-validation with 5-fold and return the trained estimators
cv_results = cross_validate(xgb_classifier,        # specify estimator 
                            X,                     # specify features
                            y.ravel(),             # specify outcome, and use ravel
                            cv=5,                  # specify 5-fold cross validation
                            return_estimator=True, # return the estimators fitted at each split
                            scoring=scoring)       # which scoring metrics to return (the whole list in this case)
 
# print the results for accuracy, recall, precision, and F1 score
for metric in ['test_accuracy', 'test_recall', 'test_precision', 'test_f1']:
    scores = cv_results[metric]
    print(f"{metric[5:]}: {scores.mean():.3f}")


```

```python
# Initialize an array to hold the feature importances
importances = np.zeros(X.shape[1])

# Average the feature importances over the folds
for estimator in cv_results['estimator']:
    importances += estimator.feature_importances_
    
# Divide by the number of folds
importances /= 5  

# Create a DataFrame for visualization
feature_importance = pd.DataFrame({'feature': X.columns, 
                                   'importance': importances})

# Sort the features by importance
feature_importance = feature_importance.sort_values('importance', ascending=False)

# Take the top 10 features
feature_importance = feature_importance.head(10)

# Plot the feature importances
plt.figure(figsize=(10, 6))

# horizontal bar plot
plt.barh(feature_importance['feature'], 
         feature_importance['importance'], 
         color='maroon', 
         align='center')

# labels
plt.xlabel('Importance')
plt.ylabel('Features')
plt.title('Feature Importance')

# gca stands for "get current axis", which allows you to modify the properties of the axes.
# and then inverts the y-axis, meaning that the values that were at the bottom will now be at the top, and vice versa.
plt.gca().invert_yaxis() 
plt.show()
```

### Model 3

```python
# create a model
logit_reg = LogisticRegression()

# fit the model
logit_model = logit_reg.fit(X_train, y_train.ravel())
```

```python
# hyperparameter tuning
# ----------

# set parameters
param_grid = {'penalty': ['l1', 'l2', 'elasticnet'],
             'C': np.arange(.1, 1, .1),
             'fit_intercept': [True, False],
             'solver': ['liblinear', 'saga']}


# execute the grid search and fit to training data
logit_grid = GridSearchCV(logit_model, 
                          param_grid, 
                          cv=3)
logit_grid.fit(X_train, 
               y_train)

# choose best performing model
best_index = np.argmax(logit_grid.cv_results_["mean_test_score"])
best_logit = logit_grid.best_estimator_
best_logit_pred = logit_grid.best_estimator_.predict(X_val)

# print results
print(logit_grid.cv_results_["params"][best_index])
print('Validation Accuracy', accuracy_score(best_logit_pred, y_val))
```

### Validation Metrics

Be sure to explain which of these metrics you would want to prioritize when conducting predictive auditing in this context and why.

**Hint**: Try writing a for loop to use [`cross_val_score()`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html) to check for accuracy, precision, recall and f1 across all of your models.

```python
models = {'XGBoost': xgb_classifier,
          'RandomForest': best_rf,
          'Logistic': best_logit}

# Loop through each model
for model_name, model in models.items():
    print(f"model: {model_name}")
    
    # Perform cross-validation
    cv_results = cross_validate(
        model, 
        X, 
        y.ravel(), 
        cv=5, 
        return_estimator=True, 
        scoring=scoring
    )
    
    # Print the results for each metric
    for metric in ['test_accuracy', 'test_recall', 'test_precision', 'test_f1']:
        scores = cv_results[metric]
        print(f"  {metric[5:]}: {scores.mean():.3f}")
    print("\n")
```

```python
# specify confusion matrix
cf_matrix = confusion_matrix(y_val, 
                             best_logit_pred, 
                             normalize = "true")

# convert to dataframe
df_cm = pd.DataFrame(cf_matrix, 
                     range(2),
                     range(2))

# plot
plt.figure(figsize = (10,7))
sns.set(font_scale=1.4)#for label size
sns.heatmap(df_cm, 
           annot=True,
           annot_kws={"size": 16},
           fmt='g')
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
