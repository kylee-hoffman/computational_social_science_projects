## Data Description and Preprocessing

For this project, you will explore data from the [National Health and Nutrition Examination Survey](https://www.kaggle.com/cdc/national-health-and-nutrition-examination-survey?select=questionnaire.csv). NHANES is a unique study that combines survey methodology with in-person medical examinations to create a dataset with demographic information, health indicators, and health outcomes.

We start by importing the data and doing some preliminary preprocessing for you. We import some libraries that will be helpful as well. 'SEQN' is the ID number for each respondent, and 'HSD010' will be our target outcome. [HSD010](https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/HSQ_H.htm#HSD010) asks for the respondent's self reported health condition, which can range from "excellent" to "poor."

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout, Input
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
sns.set_style("darkgrid")
```

```python
# Load nhanes data
nhanes = pd.read_csv('nhanes.csv')
# Get the ID numbers for each observation (seqn)
seqn = nhanes['SEQN']
# Get the target, "self-reported health condition," HSD010
hsd010 = nhanes['HSD010']
# Drop SEQN from the dataframe and then apply the standard scaler
nhanes = nhanes.drop(['SEQN', 'HSD010'], axis = 1)
nhanes_scaled = pd.DataFrame(StandardScaler().fit_transform(nhanes),
                             columns = nhanes.columns)
# Add the ID and target back in
nhanes_scaled['SEQN'] = seqn
nhanes_scaled['HSD010'] = hsd010
nhanes_scaled = nhanes_scaled.set_index('SEQN')
nhanes_scaled.head()
```

### Train PCA and Discuss Number of Components

```python
# apply PCA
# ----------
pca = PCA()
nhanes_pca = pca.fit_transform(nhanes_scaled)

# Create a Scree plot to evaluate the number of components to keep
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.xlim(0, 30)
plt.show()
```

Zooming in on the lower numbers shows that we get diminishing returns at around 6 or 7 components. 

### Barplot of Components

```python
# only plotting first 20 for better clarity
plt.bar(range(1, min(20, len(pca.explained_variance_ratio_)) + 1), pca.explained_variance_ratio_[:20])
plt.xlabel('principal component')
plt.ylabel('variance')
plt.show()
```

### How many components will you use to fit a supervised learning model?

```python
# reduce components
pca = PCA(6)
nhanes_pca = pca.fit_transform(nhanes_scaled)

plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_)
plt.xlabel('principal component')
plt.ylabel('variance')
plt.show()
```

### 2D Scatterplot of the first two components

```python
# plot first 2 components
# ----------
pca_df = pd.DataFrame(data = nhanes_pca[:, :2], 
                      columns = ['principal component 1', 
                                 'principal component 2'])

fig, ax = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

for status in y_binary.unique():
    ax[0].scatter(pca_df.iloc[(y_binary == status).values]['principal component 1'],
                  pca_df.iloc[(y_binary == status).values]['principal component 2'],
                  s=15, alpha=0.6, label=status)

ax[0].set_xlabel('Principal Component 1')
ax[0].set_ylabel('Principal Component 2')
ax[0].legend()
ax[0].set_title("binary") 

# second plot
for status in y.unique():
    ax[1].scatter(pca_df.iloc[(y == status).values]['principal component 1'],
                  pca_df.iloc[(y == status).values]['principal component 2'],
                  s=15, alpha=0.4, label=status)

ax[1].set_xlabel('Principal Component 1')
ax[1].set_ylabel('Principal Component 2')
ax[1].legend()
ax[1].set_title("all classes")

plt.show()
```

Based on the plots, it doesn't look like PCA did a very good job of separating out the different outcomes. It is a little better for the binary plot, but still not very distinct. I think this means that PCA failed to find a meaningful relationship between the features and health status.

### Choose a Clustering Algorithm

I am choosing K-means clustering. K-means is used to separate datapoints into k clusters where each datapoint is assigned to the closest cluster center.

### Cluster nhanes

```python
# elbow method to determine the optimal number of clusters
sse = []
k_values = range(1, 10)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=10)
    kmeans.fit(nhanes_scaled)
    sse.append(kmeans.inertia_)

# elbow method results
plt.plot(k_values, sse, marker='o')
plt.xlabel('# of clusters')
plt.ylabel('SSE')
plt.show()
```

Based on the plot, I think 5 is a good number of clusters.

### Plot your clusters on top of the BMI v. Income Poverty Ratio Plot

```python
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=10)
nhanes_scaled['cluster'] = kmeans.fit_predict(nhanes_scaled)

# plot BMI vs. income poverty ratio with clusters
ax = sns.scatterplot(x = "INDFMPIR", y = "BMXBMI", hue = "cluster", alpha = 0.6, palette = "tab10", data = nhanes_scaled)
ax.set(xlabel = "Income to Poverty Line Ratio",
      ylabel = "Body Mass Index")
ax.set_title("BMI v. Income-Poverty Ratio with clusters")
plt.show()
```

It looks like cluster 1 is associated with higher income, and cluster 3 is associated with lower income. Cluster 2 may be loosely assocated with lower income. Clusters 0 and 4 don't really have any clear pattern.

### Retrain the clustering algorithm on PCA components and plot clusters on your 2D scatter

```python
kmeans_pca = KMeans(n_clusters=n_clusters, random_state=10)
pca_df['cluster'] = kmeans_pca.fit_predict(pca_df.iloc[:, :2])  # only PCA components

# add cluster labels to PCA df
pca_df['cluster'] = kmeans_pca.labels_

# PCA scatterplot with clusters
for cluster in pca_df['cluster'].unique():
    cluster_data = pca_df[pca_df['cluster'] == cluster]
    plt.scatter(
        cluster_data['principal component 1'],
        cluster_data['principal component 2'],
        s=15, alpha=0.5, label=f'Cluster {cluster}')

plt.xlabel('principal component 1')
plt.ylabel('principal component 2')
plt.title('clusters on PCA components')
plt.legend()
plt.show()
```

Clustering on the first two PCA components shows a much clearer separation of the clusters compared to the previous plot. The clusters seems to have a similar number of points aside from cluster 3, which is quite sparse.

### Neural Network on Original Features

```python
# partition data
# -----------

# y = y # defined above
X = nhanes_scaled

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size = .25, 
                                                    random_state = 10)
# change labels to integers
labels = {'poor': 0, 'fair': 1, 'good': 2, 'very good': 3, 'excellent': 4}
y_train = np.array([labels[label] for label in y_train])
y_test = np.array([labels[label] for label in y_test])
```

```python
# load libraries
# -----------
import keras
from keras.utils import to_categorical

num_classes = 5
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
num_classes = y_test.shape[1]
```

```python
# create neural network model
# -----------
model = Sequential()

model.add(Input(shape=(X_train.shape[1],)))

model.add(Dense(200, kernel_initializer= 'normal', activation= 'relu')) # input layers?
model.add(Dropout(0.5))
model.add(Dense(200, activation='relu'))
model.add(Dense(num_classes, kernel_initializer= 'normal', activation= 'softmax')) # output layer?

model.compile(loss= 'categorical_crossentropy', optimizer= 'adam', metrics=['accuracy'])

model.fit(X_train,                  # specify the x data
          y_train,                  # specify the y data
          validation_data=(X_test,  # specify the validation data
                           y_test), 
          epochs=10,                # number of iterations back and forth through the network
          batch_size=10,
          verbose=2);               # print as it goes
```

### Neural Network on Principal Components

## Discussion Questions

1. In your own words, what is the difference between PCA and clustering?

PCA and clustering are quite different. PCA is primarily used in ML for dimensionality reduction, whereas clustering is used to find natural groups of datapoints based on similarity in their features. PCA outputs completely new features based on the original features, and clustering methods output clusters and does not change the original feature space.

3. How can unsupervised techniques help with downstream supervised learning tasks when working with "big data?"

Unsupervised techniques such as PCA can help when working with big data in supervised learning tasks by reducing the dimensions to help reduce overfitting or data redundancy. PCA is particular is great for reducing complexity but still retaining important information and a good amount of variance.
